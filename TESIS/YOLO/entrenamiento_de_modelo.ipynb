```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rend = pd.read_csv(direc+"results.csv")
```

```python
rend
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=80)

box = 6
plt.plot(rend['               epoch'],
         smooth(rend['      train/box_loss'],box),
         'green',
         label='Entrenamiento')
plt.plot(rend['               epoch'],
         smooth(rend['        val/box_loss'],box),
         'red',
         label='Validaci√≥n')
plt.title('')
plt.xlabel('Epocas', fontsize=20)
plt.ylabel('Box Loss', fontsize=20)
plt.grid()
plt.legend()
plt.savefig(direc+'Box_Loss.png')
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=80)

box = 6
plt.plot(rend['               epoch'],
         smooth(rend['      train/obj_loss'],box),
         'green',
         label='Entrenamiento')
plt.plot(rend['               epoch'],
         smooth(rend['        val/obj_loss'],box),
         'red',
         label='Validaci√≥n')
plt.title('')
plt.xlabel('Epocas', fontsize=20)
plt.ylabel('Object Loss', fontsize=20)
plt.grid()
plt.legend()
plt.savefig(direc+'Obj_Loss.png')
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=80)

box = 6
plt.plot(rend['               epoch'],
         smooth(rend['      train/cls_loss'],box),
         'green',
         label='Entrenamiento')
plt.plot(rend['               epoch'],
         smooth(rend['        val/cls_loss'],box),
         'red',
         label='Validaci√≥n')
plt.title('')
plt.xlabel('Epocas', fontsize=20)
plt.ylabel('Cls Loss', fontsize=20)
plt.ylim(-0.001,0.05)
plt.grid()
plt.legend()
plt.savefig(direc+'Cls_Loss.png')
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=80)

box = 16
plt.plot(rend['               epoch'],
         smooth(rend['   metrics/precision'],box),
         'k')
plt.plot(rend['               epoch'],
         rend['   metrics/precision'],
         'k',alpha=0.2)
plt.title('')
plt.xlabel('Epocas', fontsize=20)
plt.ylabel('Precisi√≥n', fontsize=20)
plt.grid()
plt.savefig(direc+'Precision.png')
```
